```python
import numpy as np
from math import pi, exp, factorial, sin, cos
import matplotlib.pyplot as plt
```

 #### 6.  Разложить функцию  y=exp(x)  в ряд Маклорена, а так же в ряд Фурье на отрезке  [−π,π] . Построить график функции и разложений.

```python
def maclaurin(xi, n):
    yi = 0
    for i in range(n): 
        yi += xi**i/factorial(i)
    return yi
```

```python
def fourier (xi, n, a0):
    fi = 0
    for i in range(1, n):
        ai = (-1)**i * (exp(pi) - exp(-pi)) / (pi * (i**2 + 1))
        bi = -i * ai
        fi += ai * cos(i * xi) + bi * sin(i * xi)
    return a0 + fi
```

```python
n = 170
x = np.linspace(-pi, pi, n)
y1 = []
y2 = []
y3 = []
a0 = (exp(pi)- exp(-pi)) / (2 * pi)
for xi in x:
    y1.append(exp(xi))    
    y2.append(maclaurin(xi, n))
    y3.append(fourier(xi, n, a0))    
    
plt.plot(x, y1, label='y = exp(x)')
plt.plot(x, y2, label='y = exp(x) - Maclaurin series')
plt.plot(x, y3, label='y = exp(x) - Fourier series')
plt.legend(bbox_to_anchor=(1, 1), bbox_transform=plt.gcf().transFigure)
plt.show()
```
