```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from math import exp
```

```python
def func(z, data):
    a, b = z
    sum_f1 = 0
    sum_f2 = 0
    for i in range(len(data)):
        x, y = data[i]
        t1 = exp(-(a * x + b))
        t2 = 1 / (1 + t1)
        f1 = -2 * (y - t2) * t2**2 * t1
        f2 = x * f1
        sum_f1 += f1
        sum_f2 += f2
    return sum_f1, sum_f2
```

```python
# Задаем рандомное распределение точек
n = 500
xi = np.random.uniform(-8, 8, n)
yi = 1 / (1 + np.exp(-np.random.uniform(0.5, 1.5, n) * xi - np.random.uniform(-1, 1, n)))

data = list(zip(xi, yi))
z0 = (0.95, 0.01)
a, b = fsolve(func, z0, data)
print(f'a = {a}, b = {b}')

y = 1 / (1 + np.exp(-(a * xi + b)))
plt.plot(xi, yi, marker="o", ls="")
plt.plot(xi, y, 'ro')
plt.show()
```
